```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_row', None)
url = "https://www.espn.com/nba/standings"

def East(url):
    df = pd.read_html(url, header=None)[0]
    df = df.reset_index().T.reset_index().T
    df = df[1].reset_index(drop = True)
    temp_df = []
    for i in range(0,len(df)):
        temp_df.append(df[i][4:].split()[1:])

    temp_df = [i for j in temp_df for i in j]
    temp_df = [elem for elem in temp_df if elem not in ["York"]]
    temp_df = pd.DataFrame(temp_df)
    df1 = pd.read_html(url, header=None)[1]
    df1
    East_standings = pd.concat([temp_df, df1], axis = "columns")
    East_standings["E/W"] = "0"
    
    return East_standings

def West(url):
    df2 =  pd.read_html(url, header=None)[2]
    df2 = df2.reset_index().T.reset_index().T
    df2 = df2[1].reset_index(drop = True)
    temp_df2 = []
    for i in range(0,len(df2)):
        temp_df2.append(df2[i][4:].split()[1:])
    

    temp_df2 = [i for j in temp_df2 for i in j]
    temp_df2 = [elem for elem in temp_df2 if elem not in ["State","Angeles","City","Orleans","Blazers","Antonio"]]
    temp_df2 = pd.DataFrame(temp_df2)
    temp_df2.loc[temp_df2[0] == "Trail"] = "Trailblazers"

    df3 = pd.read_html(url, header=None)[3]
    
    West_standings = pd.concat([temp_df2, df3], axis = "columns")
    West_standings["E/W"] = 1
    return West_standings

def Total(East, West):
    Total_standings = pd.concat([East,West])
    Total_standings.reset_index(drop = True)
    Total_standings["Team"] = Total_standings[0]
    Total_standings = Total_standings.drop(0, axis = 1)
    cols = Total_standings.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    Total_standings = Total_standings[cols]
    division_map = {"Raptors":"Atlantic", "Nets": "Atlantic", "Knicks":"Atlantic", "Pistons":"Central", "Hornets":"Southeast","76ers":"Atlantic","Celtics":"Atlantic","Pacers":"Central","Bulls":"Central","Cavaliers":"Central","Bucks":"Central","Magic":"Southeast","Wizards":"Southeast" ,"Hawks":"Southeast","Heat":"Southeast","Trailblazers":"Northwest","Jazz":"Northwest","Thunder":"Northwest","Timberwolves":"Northwest","Nuggets":"Northwest","Lakers":"Pacific","Clippers":"Pacific","Warriors":"Pacific","Suns":"Pacific","Kings":"Pacific","Rockets":"Southwest"    ,"Spurs":"Southwest"    ,"Mavericks":"Southwest"    ,"Pelicans":"Southwest", "Bobcats":"Southeast","Grizzlies":"Southwest"   }
    Total_standings["Divison"]= Total_standings["Team"].map(division_map)
 
    return Total_standings


def Standings(url):
    East_Standings = East(url)
    West_Standings = West(url)
    Total_Standings = Total(East_Standings, West_Standings)
    Total_Standings["Year"] = "2023"
    return Total_Standings

TS = Standings(url)
TS
```
